### Notes:
- **CH₄, CO₂, and N₂O variables** are grouped across percentiles (5th–95th) to reflect uncertainty bounds.
- **Aerosol forcing variables** represent both FaIR and MAGICC model outputs, covering cooling impacts.
- Selection criteria focus on:  
  - Capturing uncertainty (low/mid/high percentiles)  
  - Including multiple models for robustness  
  - Covering all major radiative forcing agents  
  - Enabling scenario flexibility (baseline, low-emission, and high-emission)
