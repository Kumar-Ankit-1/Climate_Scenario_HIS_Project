# Creating the heatmap and global map

```python

# make a safe copy to avoid SettingWithCopyWarning
df = df_filtered.copy()

# set these according to your previous results or recompute
top10_scenarios = df["Scenario"].value_counts().head(10).index.tolist()
top4_variables = df["Variable"].value_counts().head(4).index.tolist()

# choose the year to visualize (must be a column in df: 2020,2030,...)
YEAR = 2050

# filter to top scenarios and top variables
df_vis = df[
    df["Scenario"].isin(top10_scenarios) &
    df["Variable"].isin(top4_variables)
].copy()

print("shape after filter:", df_vis.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# pivot table: rows=Variable, cols=Scenario
pivot = (df_vis
         .groupby(["Variable","Scenario"])[str(YEAR)]
         .agg(np.mean)            # change aggfunc if needed
         .unstack(fill_value=np.nan)
        )

plt.figure(figsize=(14, max(4, len(pivot)*0.8)))
sns.heatmap(pivot, annot=False, cmap="viridis", linewidths=.5, linecolor="gray")
plt.title(f"Heatmap of {YEAR} values — Variables (rows) vs Scenarios (cols)")
plt.ylabel("Variable")
plt.xlabel("Scenario")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# pick scenario + variable to map
MAP_SCENARIO = top10_scenarios[0]   # or a string, e.g. "EN_NPi2020_1000"
MAP_VARIABLE = top4_variables[0]    # or a string
MAP_YEAR = YEAR                     # e.g. 2050

sub = df[
    (df["Scenario"] == MAP_SCENARIO) &
    (df["Variable"] == MAP_VARIABLE)
].copy()

# make sure region codes are uppercase
sub["Region"] = sub["Region"].str.upper()

# aggregate value per region (mean across other duplicates)
agg = sub.groupby("Region")[str(MAP_YEAR)].agg("mean").reset_index()
agg.columns = ["iso_alpha", "value"]

# drop NaNs
agg = agg.dropna(subset=["iso_alpha", "value"])

# Build choropleth. plotly expects iso_alpha as ISO-3 by default.
fig = px.choropleth(
    agg,
    locations="iso_alpha",
    color="value",
    hover_name="iso_alpha",
    color_continuous_scale="Viridis",
    title=f"{MAP_VARIABLE} — {MAP_SCENARIO} — {MAP_YEAR}"
)

fig.update_layout(margin={"r":0,"t":50,"l":0,"b":0})
fig.show()
```
